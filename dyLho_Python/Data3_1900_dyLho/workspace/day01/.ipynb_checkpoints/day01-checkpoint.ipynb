```python
 import requests

url = "https://news.naver.com/"
response = requests.get(url)
# requests 패키지에서 사용할 수 있는 .get 메소드를 활용해 url 을 불러와 response 변수에 대입

# .get() HTTP GET  요청을 보내는 메소드

print(response.status_code) # 200 : 정상 (2xx: 요청이 성공적으로 수행되었다)
print(response.text)
```

```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://news.naver.com/"
response = requests.get(url)
html = response.text
print(html)
```

```python
soup = bs(html, 'html.parser') # 응답 받은 HTML 코드를 추출할 수 있도록 변환
print(soup)
```

```python
span = soup.find('span') # 첫 번쨰 span 태그 요소

print(span)

# find 메소드는 맨 첫번째 나온 태그명 하나만 ㅏ추출하기 때문에 바로 내용만 뺄 수 있다.
```

```python
span1 = soup.find_all('span')
print(span1)

# find_all 메소드는 해당 태그명이 있는 모든 내용이 추출되기 때문에 리스트 형태로 반환되고
# 바로 내용만 뺄 수 없다 -> 내용만 빼기 위해서는 인덱스 번호를 지정해야 한다.
```

```python
print(span1[0].text)
print(span1[1].text)
print(span1[2].text)
print(span1[3].text)
```

```python
# 언론사별 이라는 문자만 추출하기
test = soup.find('span', class_ = 'Nitem_link_menu').text

print(test)
```

```python
tests = soup.find_all('span', class_ = "Nitem_link_menu")
print(tests)
print("\ntests[3].text:: ",tests[3].text)
```

## 야구 순위, 구단명 추출하기
### https://sports.news.naver.com/kbaseball/record/index?category=kbo

```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://sports.news.naver.com/kbaseball/record/index?category=kbo"

# 1. 상태코드 확인
resonse = requests.get(url)
print("resonse.status_code:: ", resonse.status_code)
print("type??:: ", type(resonse))
```

```python
# 2. html 코드 불러오기
html = resonse.text
print(html)
```

```python
import requests
from bs4 import BeautifulSoup as bs

soup = bs(html, 'html.parser')
print(soup)
```

```python
# 팀순위라는 문구 추출
# find : 맨 처음 것 하나만 추출, .text메소드 바로 사용 가능
text1 = soup.find('h4')
text2 = soup.find('h4').text
print(text1)
print(text2)

# find_all : 리스트 형태로 추출, .text매소드를 사용하기 위해서는 인덱싱을 이용해 해당 인덱스 지정
data1 = soup.find_all('strong')

print(data1)
```

```python
data2 = soup.find_all('strong')[4].text
print(data2)
```

```python
# 순위표 가져오기
ranking_table = soup.find('tbody', id = 'regularTeamRecordList_table')
print(ranking_table)
```

```python
row = ranking_table.find_all('tr')
print(row)
```

```python
for i in row:
    # 순위
    th = i.find('th').text.strip()
    # print(th)

    # 구단명
    div = i.find('div').text.strip()
    # print(div)

    print(f"{th}위 {div}")
    
# strip() 문자열 메서드: 양쪽의 불필요한 공백 제거 메서드
```

```python
# 추출한 순위 메모장에 저장하기
ranking_table = soup.find('tbody', id = 'regularTeamRecordList_table')


# 1. 파일 작성
with open("./baseball_ranking.txt", "w", encoding='utf-8') as f: # 파일 객체를 f 라고 하겠다.
    row = ranking_table.find_all('tr')

    for i in row:
        th = i.find('th').text.strip()
        div = i.find('div').text.strip()
        f.write(f"{th}위 {div}\n")


print("순위와 구단명을 파일에 저장했습니다.")
```

## 기사 크롤링
### https://www.spotvnews.co.kr/news/articleView.html?idxno=677558#_PA

## 기사 제목 
## 기사 내용
## 메모장에 저장하기

```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://www.spotvnews.co.kr/news/articleView.html?idxno=677558#_PA"

# 1. 상태코드 확인
resonse = requests.get(url)
print("resonse.status_code:: ", resonse.status_code)
resonse.encoding = 'utf-8'
html = resonse.text
#print(html)
```

```python
soup = bs(html, 'html.parser')
#print(soup)
```

```python
# 기사제목

title = soup.find('h3').text
print(title)
```

```python
# 기사 내용

content = soup.find_all('p')

print(content)
```

```python
# 1. 파일 작성
with open("./football_article.txt", "w", encoding='utf-8') as f: # 파일 객체를 f 라고 하겠다.
    contents = soup.find_all('p')

    f.write(f"[제목] {title}\n")

    for i in contents:
        content = i.text.strip()
        f.write(f"\n {content}\n")

print("제목과 내용을 파일에 저장했습니다.")
```
