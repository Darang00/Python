```python
import pandas as pd

# 주어진 파일 경로
file_path = 'D:/dyLho_Python/Data3_1900_dyLho/project/불법주정차_신고현황_동대문구_샘플추출_26.csv'

# CSV 파일 읽기
df = pd.read_csv(file_path)

# '도로명주소' 컬럼에서 동네 데이터 추출
df['동네'] = df['도로명주소'].apply(lambda x: x.split()[-3])

# 동네별 데이터 수 구하기
neighborhood_counts = df['동네'].value_counts()

print("동네별 데이터 수:")
print(neighborhood_counts)
```

```python
# 주어진 파일 경로
file_path = 'D:/dyLho_Python/Data3_1900_dyLho/project/불법주정차_신고현황_동대문구_샘플추출_26.csv'

# CSV 파일 읽기
df = pd.read_csv(file_path)

# '민원접수일' 컬럼을 날짜 형식으로 변환
df['민원접수일'] = pd.to_datetime(df['민원접수일'], format='%Y%m%d')

# 월별 데이터 건수 구하기
monthly_counts = df['민원접수일'].dt.month.value_counts().sort_index()

print("월별 데이터 건수:")
print(monthly_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# pip install matplotlib

# 주어진 파일 경로
file_path = 'D:/dyLho_Python/Data3_1900_dyLho/project/불법주정차_신고현황_동대문구_샘플추출_26.csv'

# CSV 파일 읽기
df = pd.read_csv(file_path)

# '민원접수일' 컬럼을 날짜 형식으로 변환
df['민원접수일'] = pd.to_datetime(df['민원접수일'], format='%Y%m%d')

# 월별 데이터 건수 구하기
monthly_counts = df['민원접수일'].dt.month.value_counts().sort_index()

# 꺾은선 그래프로 시각화
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='line', marker='o', color='b')
plt.title('월별 데이터 건수')
plt.xlabel('월')
plt.ylabel('건수')
plt.grid(True)
plt.xticks(range(1, 13))
plt.show()
```
