```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# 나이가 10대(10~19)인 승객만 선택
# age >= 10 and(&) age < 20
sr_age = (df['age'] >= 10) & (df['age'] < 20)
print(sr_age, type(sr_age)) #논리형 값으로 구성된 시리즈 반환

df_ten = df.loc[sr_age]
df_ten.head()
```

```python
# <실습>
# 1) 나이가 10살 미만이고 성별이 여성인 승객이 age, sex, alive 열만 추출
# age < 10 & sex == 'female'
sr = (df['age'] < 10) & (df['sex'] == 'female')
df_female = df.loc[sr, ['age', 'sex', 'alive']]

df_female.head()

```

```python
# 성별이 남성이거나 나이가 15살 미만인 승객 중 생존한 사람들의 인원 수 구하기
# df['age'] < 15 | (or) df['sex'] = 'male' & df['survived'] == 1


sr = ((df['age'] < 15) | (df['sex'] == 'male')) & (df['survived'] == 1)

df_survived = df.loc[sr]

print(f"남성이거나 15살 미만인 승객 중 생존자는 : {len(df_survived)}명")

df_survived
```

```python
# 현재 탑승한 형제 또는 배우자의 수가 3, 4, 5 인 승객만 따로 추출
# sibsp: 함께 탑승한 형제 및 배우자의 수

# 1) boolean 인덱싱
sr = (df['sibsp'] >= 3) & (df['sibsp'] <= 5)

df_si = df.loc[sr]

df_si.head()
```

```python
# isin() 메소드를 사용하여 동일한 조건으로 데이터 추출
isin_sr = df['sibsp'].isin([3, 4, 5])

print(isin_sr) # 시리즈 형태로 반환

df_isin = df.loc[isin_sr]

df_isin.head()
```

## 데이터 프레임 합치기


### 1) 데이터 프레임 연결

```python
# 데이터 프레임 연결: pd.concat(데이터프레임 리스트)

# 데이터 프레임 생성
df1 = pd.DataFrame({"a":[f"a{x}" for x in range(4)],
                   "b": [f"b{x}" for x in range(4)],
                   "c": [f"c{x}" for x in range(4)]
                   },
                  index = ["zero", "one", "two", "three"]
                  )

print(df1, "\n")
```

```python
df2 = pd.DataFrame({"a": [f"a{x}" for x in range(2, 6)],
                   "b": [f"b{x}" for x in range(2, 6)],
                   "c": [f"c{x}" for x in range(2, 6)],
                   "d": [f"d{x}" for x in range(2, 6)]
                   },
                  index = ['two', 'three', 'four', 'five']
                  )

print(df2, "\n")
```

```python
# axis = 0(디폴트): 2개의 데이터 프레임 방향으로 이어 붙여 연결

result1 = pd.concat([df1, df2])

result1 # df1 아래에 df2가 연결됨
# df1은 d열이 없으므로 NaN 값으로 채워짐
```

```python
# ignore_index = True 옵셥: 새로운 행 인덱스를 설정, 정수형 인덱스

result2 = pd.concat([df1, df2], ignore_index = True)


print(result2) # 행 인덱스(0 ~ 7): 정수형 인덱스
```

```python
print(df1, "\n")
print(df2, "\n")

# axis = 1 : 열 방향으로 (옆으로) 데이터 프레임 연결
result3 = pd.concat([df1, df2], axis = 1)
result3
# zero, one 행은 df1 에만 존재, four, five 행은 df2에만 존재한다.
# 어느 한쪽에만 데이터가 존재할 경우, 비어있는 위치의 값은 NaN으로 설정
```

```python
# join = 'inner': 연결할 데이터프레임들의 행이나 열 인덱스의 교집합 기준으로 연결
print(df1, "\n")
print(df2, "\n")

result4 = pd.concat([df1, df2], join='inner')
print(result4, "\n") # 행 방향으로 연결할 때, df1, df2 의 공통되는 열 인덱스 a, b, c 기준으로 연결


result5 = pd.concat([df1, df2], axis = 1, join = "inner")
print(result5) # 열 방향으로 연결할 때, df1, df2 의 공통되는 행 인덱스 two, three 행 기준으로 연결
```

### 2) 데이터프레임 병합

```python
import pandas as pd
# 데이터프레임 병합: pd.merge(df_left, df_right, how='inner', on=None)

# 데이터프레임 생성
df_left = pd.DataFrame({"번호": [1, 2, 3, 4],
                       "이름": ["짱구", "철수", "유리", "맹구"],
                        "나이": [5, 6, 7, 6]
                       })

print(df_left, "\n")

df_right = pd.DataFrame({"번호": [1, 3, 5, 6],
                        "학생이름": ["짱구", "유리", "둘리", "또치"],
                         "점수": [60, 85, 70, 90]
                        })

print(df_right, "\n")

result1 = pd.merge(df_left, df_right)
# df_left, df_right 에 공통으로 있는 '번호' 열 기준으로 병합: on = None(디폴트)
# 기준이 되는 열(번호)의 데이터가 양 쪽에 같이 존재하는 경우(1, 3) 출력: how = "inner" (디폴트)
print(result1, "\n")
```

```python
print(df_left, "\n")
print(df_right, "\n")

# 데이터프레임 병합: how = "outer"
# 번호 열 기준(key)으로 병합
result2= pd.merge(df_left, df_right, how='outer', on='번호')

print(result2)
```

```python
# 데이터프레임 병합: 왼쪽 데이터 기준으로 병합 -> how="left"
print(df_left, "\n")
print(df_right, "\n")

result3 = pd.merge(df_left, df_right, how="left")
print(result3, "\n")

result4 = pd.merge(df_left, df_right, how="left", left_on="이름", right_on ='학생이름')
print(result4) # 왼쪽의 이름 열과 오른쪽의 학생이름 열을 키열로 설정(왼쪽 데이터 기준으로 병합)
# 번호 열이 두 데이터 프레임에 모두 존재하므로 '번호_x', '번호_y'와 같이 구분되어서 표시됨
```

```python
# 데이터 프레임 합치기: 오른쪽 데이터 기준 how='right'

print(df_left, "\n")
print(df_right, "\n")

result5 = pd.merge(df_left, df_right, how='right')
print(result5) # 오른쪽 데이터의 번호열 기준으로 병합

```

```python
result6 = pd.merge(df_left, df_right, how="right", left_on="이름", right_on="학생이름")
print(result6) # 오른쪽 데이터프레임 키열(학생이름) 데이터 기준으로 병합
```

```python
# merge() 함수와 boolean 인덱스 활용 데이터 조회
# 나이가 7살 미만인 학생의 점수 확인
print(df_left['나이'] < 7, "\n")

under_seven = df_left[df_left['나이'] < 7]
print("under_seven\n",under_seven)
print("\n type??", type(under_seven)) # pandas.core.frame.DataFrame
```

```python
print("under_seven\n",under_seven)

print("\n",df_right, "\n")

result = pd.merge(under_seven, df_right)
print(result)
```
