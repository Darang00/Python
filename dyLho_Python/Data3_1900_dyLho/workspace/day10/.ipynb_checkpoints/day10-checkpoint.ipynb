## 분류

```python
# titanic 데이터셋의 데이터로 탑승객의 생존여부를 예측(분류 알고리즘)

# step1 - 데이터 준비

import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

df.head()
```

```python
# 결측치가 많은 deck열 삭제, embarked와 내용이 겹치는 embark_town 열 살제
rdf = df.drop(['deck', 'embark_town'], axis = 1)

# age열의 값이 결측치인 행 제거
rdf = rdf.dropna(subset = ['age'], how = 'any', axis = 0)

rdf.head()
```

```python
# embark열의 결측치 값을 승선도시 중에서 많이 출현한 값으로 대체
# embark열의 최빈값
most_freq = rdf['embarked'].value_counts(dropna = True).idxmax()
print(f"embarkded열의 최빈값 : {most_freq}", "\n")

# print(rdf.describe(include = 'all'))
print("\n")

# 결측치를 최빈값으로 치환
rdf['embarked'].fillna(most_freq, inplace = True)

print(rdf.isnull().sum())
```

```python
# 결측치가 많은 deck열 삭제, embarked와 내용이 겹치는 embark_town 열 살제
rdf = df.drop(['deck', 'embark_town'], axis = 1)

# age열의 값이 결측치인 행 제거
rdf = rdf.dropna(subset = ['age'], how = 'any', axis = 0)

rdf.head()
```

```python
# embark열의 결측치 값을 승선도시 중에서 많이 출현한 값으로 대체
# embark열의 최빈값

most_freq = rdf['embarked'].value_counts(dropna = True).idxmax()
print(f"embarkded열의 최빈값: {most_freq}", "\n")

print(rdf.describe(include = 'all'))
print('\n')

# 결측치를 최빈값으로 치환
rdf['embarked'].fillna(most_freq, inplace = True)

print(rdf.isnull().sum())
```

```python
# step3 - 속성 선택()

ndf = rdf[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'embarked']]

ndf.head()
```

```python
# 범주형 데이터(sex, embarked)를 모형이 인식할 수 있도록 숫자형(더미변수)로 변환
# 성별 데이터 더미변수로 변환
# pd.get_dummies(df['col'])
dummy_sex = pd.get_dummies(ndf['sex']).astype('int')
print(dummy_sex)

# 기존의 데이터프레임에 연결
# pd.concat([df1, df2])
ndf = pd.concat([ndf, dummy_sex], axis = 1)

ndf.head()
```

```python
# 승선 데이터 더미 변수로 변환
dummy_emb = pd.get_dummies(ndf['embarked'], prefix = 'town').astype('int')
# prefix옵션 : 열 이름에 접두어(town)추가
# print(dummy_emb)

# 데이터프레임에 연결
ndf = pd.concat([ndf, dummy_emb], axis = 1)

ndf.head()
```

```python
# 기존의 sex, embarked 열 제거
ndf.drop(['sex', 'embarked'], axis = 1, inplace = True)

ndf.head()
```

```python
# step4 - 훈련/검증 데이터 분할

# 독립변수(설명변수)
x = ndf.loc[:, 'pclass':'town_S']
# 종속변수(목표값, 예측변수)
y = ndf['survived']

# 훈련, 검증 데이터 분할
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 10)

print(f"train data : {x_train.shape}")
print(f"test data : {x_test.shape}")
```

```python
# step5 - 모형 학습 및 검증

# KNN분류 모형 객체 생성
from sklearn.neighbors import KNeighborsClassifier

# 모형 객체 생성
knn = KNeighborsClassifier(n_neighbors = 5)
# n_nighbors : k(이웃 개수)값 설정

# 훈련 데이터로 모형 학습
knn.fit(x_train, y_train)

# 검증 데이터로 survived(생존자) 예측
y_pre = knn.predict(x_test)

print(y_pre[:10])
print(y_test.values[:10])

# 예측 확률
print(knn.score(x_test, y_test))
```

```python
# 모형 성능 평가 - Confusion Matrix 계산

from sklearn import metrics

knn_metrix = metrics.confusion_matrix(y_test, y_pre)
print(knn_metrix) #[[TN, FP], [FN, TP]]
```

```python
# 평가지표 계산
report = metrics.classification_report(y_test, y_pre)
print(report)
# f1 - score : 생존(1) 예측의 정확도 0.72, 미생존(0)의 예측 정확도는 0.82
# 예측 능력에 차이는 있지만 비교적 괜찮은 예측력을 가지고 있다
```

```python
# KNN 분석 모형에서 알맞는 K값(이웃의 개수)을 설정하는 것이 중요
# 그래프로 분류모형에 적절한 k값 확인
import matplotlib.pyplot as plt

# 예측 확률을 저장할 리스트
test_accuracy = []

for n_neighbors in range(1, 11) :
    # 모델생성
    knn = KNeighborsClassifier(n_neighbors = n_neighbors)
    knn.fit(x_train, y_train)
    # 정확도 계산, 리스트에 추가
    test_accuracy.append(knn.score(x_train, y_train))


#print(test_accaracy)
fig = plt.figure(figsize = (10, 5))
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.plot(range(1, 11), test_accuracy, marker = 'o', label = '예측 정확도')
plt.ylabel('정확도')
plt.xlabel('n_neighbors')
plt.legend()
plt.show()


# k값이 커지면 모델은 단순해지고 K값이 적을수록 모델이 복잡해진다
#  -> k값이 커질수록 모델은 단순해지지만 정확도는 떨어진다
# k값이 1이면 이웃이 1개이기 때문에 예측이 정확하다
# 하지만 분류기준이 명확해지지 않는다
# 따라서 중간 값인 5, 6정도가 해당 분류 모델에 사용하기 적합하다
```

## 군집 알고리즘

```python
import seaborn as sns

iris = sns.load_dataset('iris')

# sepal_length: 꽃받침의 길이 (단위: cm)
# sepal_width: 꽃받침의 너비 (단위: cm)
# petal_length: 꽃잎의 길이 (단위: cm)
# petal_width: 꽃잎의 너비 (단위: cm)
# species: 붓꽃의 품종
# 품종의 종류(3가지):
# setosa: 세토사 종
# versicolor: 베르시컬러 종
# virginica: 버지니카 종

iris.head()
```

```python
# 데이터 자료형 확인
# 데이터 탐색 -> dataset.info()

print(iris.info())

print("\n")

# 통계 요약 정보 확인
print(iris.describe())
```

```python
# 분석에 사용할 속성 선택
# 꽃잎, 꽃받침의 길이, 너비 -> 군집분석

x = iris.loc[:, :'petal_width'] # 행은 전체, 열은 petal_width 까지

x.head()
```

```python
# k-means 군집모형
from sklearn import cluster

# 모형 객체 생성
kmeans = cluster.KMeans(init = 'k-means++', n_clusters = 3, n_init = 10)
# n_clusters = 3 :집단 3개 만들거다.  (k(클러스터 개수) 값 설정 (0~2))
# n_init: 초기 중심위치 찾기 시도 횟수, 디폴트는 10, 10개의 무작위  중심위치 후보 중 가장 좋은 값을 선택

# 모형 학습
kmeans.fit(x) # 비지도 학습 (정답을 알려주지 않고 독립변수만 알려주고서 알아서 찾도록 함)

# 예측(군집)
cluster_label = kmeans.labels_ 
# print(cluster_label)

# 예측 결과를 데이터 프레임에 추가
iris['cluster'] = cluster_label
iris.head()
```

```python
# 그래프로 분석 평가/검증
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 군집분석(예측) 산점도 그래프: 클러스터별 꽃받침의 길이, 너비
groups_cluster = iris.groupby('cluster')

for key, group in groups_cluster:
    ax1.scatter(group['sepal_length'], group['sepal_width'], label = f'cluster{key}')


ax1.set_xlabel('꽃받침의 길이')
ax1.set_ylabel('꽃받침의 너비')
ax1.set_title('군집예측')
ax1.legend()

# 실제 데이터 산점도 그래프: 붓꽃의 종류별 꽃받침의 길이, 너비
groups_species = iris.groupby('species')

for key, group in groups_species:
    ax2.scatter(group['sepal_length'], group['sepal_width'], label = key)

ax2.set_xlabel('꽃받침의 길이')
ax2.set_ylabel('꽃받침의 너비')
ax2.set_title('군집예측')
ax2.legend()

plt.show()
```
