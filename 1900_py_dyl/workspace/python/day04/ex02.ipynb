```python
import pandas as pd   

df_park = pd.read_csv('서울시 주요 공원현황.csv', encoding='euc-kr')

pd.set_option('display.max_colwidth', 10) # 대략 한 컬럼 당 10글자 정도

df_park

#df_park.info()

# 필요없는 컬럼들 드랍시키기
df_park.drop(['면적', '개원일', '주요식물', '주요시설', '이용시참고사항', 'X좌표(GRS80TM)', 'Y좌표(GRS80TM)'], axis=1, inplace=True)

df_park.head(3)

# 컬럼 이름 재정의
df_park.rename(columns = {'X좌표(WGS84)':'경도', 'Y좌표(WGS84)':'위도'}, inplace=True)
df_park.head(3)

```

```python
import folium

# 위도
lat = 37.5501402

# 경도
lng = 126.9903773

map = folium.Map(
    location = [lat, lng], # 최초 로딩 시 지도가 보여줄 위치를 [위도, 경도]로 설정 
    width = 800,           # 지도의 너비
    height = 600,          # 지도의 높이
    zoom_start = 15        # 줌은 최대 18까지 설정 가능 
)

map
```

```python
# 지도에 마커(핑) 추가하기 - 공식 홈페이지 Marker 들어가서 보면 어케 쓰는지 나옴
marker = folium.Marker(
    location = [lat, lng],             # 핑 찍을 위치를 Marker 객체에 넣음(marker의 위치지정)
    popup='남산공원',       # 마커 클릭 시 나오는 팝업창
    tooltip='툴팁이다!',    # 마커에 마우스 올렸을 때 나타나는 툴팁
    icon=folium.Icon('red') # 마커 아이콘 스타일 변경
).add_to(map) # 핑을 어디에 적용할건지 -> map (지도에 마커 추가)

map
```

```python
df_loc = df_park[['위도', '경도', '공원명', '공원개요', '이미지']].dropna()

for sub_lat, sub_lng, sub_name, sub_info, sub_img in zip(df_loc['위도'], df_loc['경도'], df_loc['공원명'], df_loc['공원개요'], df_loc['이미지']):
    #print(sub_lat, sub_lng)
    folium.Marker(
        [sub_lat, sub_lng],
        tooltip = sub_name,
        popup = f'<div style="width:300px">{sub_info}</div> <img src="{sub_img}" style="width:300px">'
    ).add_to(map)

map

map.save('test.html')
```
