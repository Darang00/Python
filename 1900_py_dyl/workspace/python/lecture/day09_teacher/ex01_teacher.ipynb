```python
from sklearn.model_selection import train_test_split

X_data = df.loc[:, 'sepal_length':'petal_width']
y_data = df.loc[:, 'target']

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=47)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

KNN 모델

```python
'''
K 최근접 이웃 모델

예측하려는 데이터 X가 주어지면 기존 데이터 속성 중 속성이 비슷한 K개의 이웃을 먼저 찾는 알고리즘을 사용한다.
새로운 데이터와 가장 가까운 이웃 데이터를 K개 찾고 결과를 도출한다.
직선 거리상 가장 가까운 이웃데이터의 값이 결과로 사용된다.
'''

from sklearn.neighbors import KNeighborsClassifier as kn

# n_neighbors=7 은 가장 가까운 7개의 이웃을 보고 어느 분류에 속하는지 결정하겠다는 뜻
knn = kn(n_neighbors=7)

# 훈련
knn.fit(X_train, y_train)

# 테스트
y_knn_pred = knn.predict(X_test)

print("예측값 : ", y_knn_pred[:10])
print("실제값 : ", y_test[:10].values)
```

```python
from sklearn.metrics import accuracy_score

# accuracy_score() 함수는 정확도를 측정하는 함수이다.
knn_acc = accuracy_score(y_test, y_knn_pred)
print('정확도 : ', knn_acc)
```
