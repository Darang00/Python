```python
'''
동대문구 내 지역별 인구 데이터
https://www.ddm.go.kr/ddmmap/selectPopltnInfoWebList.do?key=1650


서울 불법주정차 현황
https://data.seoul.go.kr/dataList/OA-22189/F/1/datasetView.do
'''
import pandas as pd

# Load the data from the uploaded CSV file
file_path = 'D:/dyLho_Python/Data3_1900_dyLho/project/불법주정차 신고현황(23년11월1일_24년3월13일).csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()

```

```python
from geopy.geocoders import Nominatim
import pandas as pd

geolocator = Nominatim(user_agent='South Korea', timeout=None)

# 데이터를 더 작은 크기로 줄임 (예: 1000개 샘플)
sample_data = data.sample(n=1000, random_state=1)

# 도로명주소를 저장할 리스트
addresses = []

for index, row in sample_data.iterrows():
    try:
        location = geolocator.reverse((row['위도'], row['경도']), exactly_one=True)
        address = location.address
    except Exception as e:
        address = None
        print(f"Error: {e} at index {index}")
    addresses.append(address)
    print(index, address)

# 데이터프레임에 도로명주소 컬럼 추가
sample_data['도로명주소'] = addresses

```

```python
sample_data.head()
```

```python
# 파일로 저장
sample_output_file_path = 'D:/dyLho_Python/Data3_1900_dyLho/project/불법주정차_신고현황_샘플_도로명주소_추가.csv'
sample_data.to_csv(sample_output_file_path, index=False)

sample_output_file_path
```

```python
# '도로명주소' 컬럼을 가공하는 함수 정의
def process_address(address):
    if pd.isnull(address):
        return address
    
    # ',' 문자 제거
    address = address.replace(',', '')
    
    # 마지막 두 단어 제거
    address_parts = address.split()[:-2]
    
    # 다시 문자열로 결합
    return ' '.join(address_parts)

# 샘플 데이터를 만들어 가공
#sample_data = data.head(10)  # 예시로 10개 데이터만 사용

# '도로명주소' 컬럼에 적용
processed_addresses = [process_address(addr) for addr in sample_data['도로명주소']]

# 결과 확인
processed_addresses
```

```python
# CSV 파일 읽기
df = pd.read_csv(sample_output_file_path)
df['도로명주소'] = processed_addresses

# 수정된 데이터를 CSV 파일로 저장
df.to_csv(file_path, index=False)

df.head(15)
```

```python
# '도로명주소' 컬럼에서 '동대문구'를 포함하는 행 추출
filtered_df = df[df['도로명주소'].str.contains('동대문구')]

# 새로운 CSV 파일로 저장
filtered_file_path = 'D:/dyLho_Python/Data3_1900_dyLho/project/불법주정차_신고현황_동대문구_추출.csv'
filtered_df.to_csv(filtered_file_path, index=False)

filtered_df.head(15)
```
