```python
import pandas as pd

df = pd.read_csv("./online_retail.csv")

# print(df.dtypes)
df['new_date'] = pd.to_datetime(df['InvoiceDate'])
print(df['new_date'].dtype)

df.drop('InvoiceDate', axis = 1, inplace = True)

df
```

```python
# 연, 월, 일 따로 구분 : dt속성을 이용
df['Year'] = df['new_date'].dt.year
df['Month'] = df['new_date'].dt.month
df['Day'] = df['new_date'].dt.day

df
```

```python
# Timestamp를 Period 로 변환하여 연, 월, 일 분리
df['Date_Year'] = df['new_date'].dt.to_period(freq = "A") # 연도기준
df['Date_Month'] = df['new_date'].dt.to_period(freq = "M")

df
```

```python
# 연-월 데이터를 index로 지정
df.set_index('Date_Month', inplace = True) 

df
```

```python
# 날짜 사이의 간격 계산

open_day = pd.to_datetime("2010-12-01")
print(open_day, type(open_day))

# DateFrame에 있는 인덱스 datetime 형식으로 변환
df.index = df.index.to_timestamp() #Period객체를 TimeStamp로 변환

df
```

```python
# time_lapse 계산
df['time_lapse'] = df.index - open_day

df
```

## 함수 매핑

## 1) 시리즈 객체 매핑

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# 사용자 정의 함수 활용
# 함수 정의

def add_ten(element) :
    return element + 10

def add_two(x, y) :
    return x + y

# 함수 호출
# print(add_ten(5))
# print(add_two(10, 20))

# print(df['age'], type(df['age']))

# 시리즈 원소에 함수 매핑 : sr.apply(함수)
sr1 = df['age'].apply(add_ten) #age열(시리즈 형태) 함수 적용, element : df['age']의 모든 원소
# print(sr1.head(), type(sr1))

sr2 = df['age'].apply(add_two, y = 5) # x : df['age']의 모든 원소, y : 5
# print(sr2.head())

# lambda 함수 활용
sr3 = df['age'].apply(lambda x : x + 10) # x : df['age']의 데이터
print(sr3.head())

df.head()
```

```python
# 2) 시리즈 객체에 함수 매핑

# 결측치 확인 사용자 정의 함수
def is_missing(sr) : # 시리즈를 인수로 받음
    return sr.isnull()  # 결측치 확인 후 시리즈로 반환

df = df[['age', 'fare']]
# (1) 데이터 프레임의 각 열(시리즈)에 함수 적용
result = df.apply(is_missing, axis=0) #age, fare열에 각각 적용됨
result = df.apply(lambda sr : sr.isnull())  #람다로 적용 sr : 데이터 프레임 열이 들어감
print(result.head(), type(result)) 


df.head()
```

```python
# (2) 데이터 프레임의 각 행(시리즈)에 함수 적용

df['result1'] = df.apply(lambda x : x['age'] + x['fare'], axis = 1) # x : 데이터프레임의 각 행
df['result2'] = df.apply(lambda x : add_two(x['age'], x['fare']), axis = 1)
df.head()
```

### 2) 데이터 프레임 함수 매핑

```python
# (1) 데이터 프레임 원소에 함수 매핑
# df.map(함수)

df = sns.load_dataset('titanic')
df = df[['age', 'fare']]


df_map = df.map(add_ten)

print(df_map.head(), type(df_map))

df.head()
```

```python
df_map = df.map(lambda x : x + 10) # x : 데이터프레임 원소

print(df_map.head(), type(df_map))

df.head()
```

```python
# 2) 데이터 프레임 객첵에 함수 매핑

# df.pipe(함수)

# 사용자 정의 함수
# 1) 각 열의 NaN 찾기: 데이터 프레임을 전달하면 데이터 프레임 반환
def null_value(df):
    return df.isnull()


# 2) 각 열의 NaN 개수 반환: 데이터 프레임을 전달하면 시리즈로 반환
def null_count(df):
    return df.isnull()


# 3) 데이터 프레임의 총 NaN 개수: 데이터프레임을 전달하면 값을 반환
def total_null(df):
    return df.isnull().sum().sum()


# print(df.isnull().sum().sum(), type(df.isnull().sum().sum()))

result_df = df.pipe(null_value) # 데이터프레임 반환하는 함수 적용
result_sr = df.pipe(null_count) # 시리즈를 반환하는 함수 적용
result_value = df.pipe(total_null) # 시리즈에서 열마다의 결측치의 개수를 합침, 값으로 반환

print("===== 데이터 프레임 반환 =====")
print(result_df.head(), type(result_df))
print("===== 시리즈 반환 =====")
print(result_sr.head(), type(result_sr))
print("===== 특정 값 반환 =====")
print(result_value, type(result_value))

df.head()
```

# 열 분리

```python
df = pd.read_csv("./online_retail.csv")

df = df.loc[:10, "InvoiceNo" : 'InvoiceDate']

df
```

```python
# 열 분리
# 데이터 프레임 데이터 타입 확인
print(df.dtypes)


# InvoiceDate 열 데이터 날짜와 시간으로 분리
# .split() : 문자열을 리스트 변환
df_date = df['InvoiceDate'].str.split(" ")
print(df_date)
```

```python
# split()
str1 = "a b c d"


print(str1.split(" "))
```

```python
# 날짜와 시간 따로 선택
# sr.str.get(idx): 시리즈의 문자열 리스트 인덱싱
print(df_date.str.get(0)) # 문자열 리스트 첫번째 요소 선택

# 데이터 프레임에 날짜와 시간 열 추가
df['date'] = df_date.str.get(0)
df['time'] = df_date.str.get(1)
df
```

```python
# <실습>
# 1) date 열을 year(연도), month(달), day(일) 3개의 열로 분리

df_datecol = df['date'].str.split("/")
df['year'] = df_datecol.str.get(2) + "년"
df['month'] = df_datecol.str.get(0) + "월"
df['day'] = df_datecol.str.get(1) + "일"

df
```
