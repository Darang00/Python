## 데이터 개수 확인

## 판다스 통계함수

```python
import pandas as pd

# 데이터 프레임 생성
df = pd.DataFrame({"수학": [80, 75, 90],
                   "영어": [90, 60, 70],
                   "국어": [75, 90, 85]                 
                  }, index = [f"학생{x}" for x in range(1, 4)])

df
```

```python
# 과목마다 평균 점수
print(df.mean(), type(df.mean()))
print(df.mean().round(2), type(df.mean())) # 각 열마다의 평균을 반환, 시리즈 형태
print(df.mean().astype('int'), type(df.mean()))

print(type(df.mean().values)) # 넘파이 배열 형태

# 수학의 평균 점수
print(f"수학 평균 점수: {df['수학'].mean()}점")

# 가장 높은 국어 점수
print(f"국어 최대 점수: {df['국어'].max()}점")

# 가장 낮은 수학 점수
print(f"수학 최소 점수: {df['수학'].min()}점")

# 수학과 국어 점수의 중간 점수
print("\nmedian()\n",df[['수학', '국어']].median())

# 과목별 합계 점수
print("\ndf.sum()\n",df.sum())

# 각 과목의 표준편차
print("\ndf.std()\n", df.std())

```

## Matplotlib
### 1) 꺾은선 그래프

```python
# matplotlib
import matplotlib.pyplot as plt

# 영역 설정
plt.figure(figsize = (8, 5)) # 그래프 크기 설정
plt.rcParams['font.family'] = 'Malgun Gothic' # 폰트 설정(한글 깨짐 방지)

x = [f"{x}월" for x in range(1, 6)]
# print(x)
y = [2, 5, 7, 2, 1]

# 꺾은선 그래프: plot(x, y)
plt.plot(x, y, color = 'darkblue', marker = 'o', markerfacecolor='green', linestyle='dashed')

plt.title('월별 인원수', fontsize = 20) # 그래프 제목
plt.xlabel("월", fontsize = 16) # x축 라벨
plt.ylabel("인원수") # y축 라벨


plt.show()
```

## 2) 막대 그래프

```python
import matplotlib.pyplot as plt

x = ['봄', '여름', '가을', '겨울']
y = [250, 100, 230, 150]
plt.bar(x, y, color = ['green', 'red', 'brown', 'gold']) # 색깔 따로 지정할 수 있음

plt.xlabel("계절")
plt.ylabel("매출액")
plt.title('계절별 매출 통계')
plt.show()
```

## 그래프 여러개 그리기

```python
import matplotlib.pyplot as plt

x1 = [1, 3, 5, 7] # x축의 데이터 값
y1 = [1, 2, 3, 4] # y축의 데이터 값

plt.bar(x1, y1, color='r')
plt.show()


# ------------------------------------------------

plt.bar(x1, y1, color = 'b', width = 0.5)
plt.show()

# ------------------------------------------------

plt.bar(x1, y1, color='g', width = 0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 8))

x1 = [1, 3, 5, 7] # x축의 데이터 값
y1 = [1, 2, 3, 4] # y축의 데이터 값

axes1 = fig.add_subplot(121) # 1행 2열의 1번째
axes1.bar(x1, y1, color = 'r')


# ------------------------------------------------

axes2 = fig.add_subplot(222) # 2행 2열의 2번째
axes2.bar(x1, y1, color = 'b', width = 0.5)

# ------------------------------------------------

axes3 = fig.add_subplot(224) # 2행 2열의 4번째
axes3.bar(x1, y1, color = 'g', width = 0.7)

plt.show()
```
