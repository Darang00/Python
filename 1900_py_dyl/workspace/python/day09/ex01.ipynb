```python
from sklearn.model_selection import train_test_split 

X_data = df.loc[:, 'sepal_length':'petal_width']
y_data = df.loc[:, 'target']


X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.25, random_state=80) # random_state를 여러번 바꾸면서 정확도가 어느정도인지 확인해 봐야한다.

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## KNN 모델

```python
# KNN 모델: 분류할 때 사용하는 모델
'''
K 최근접 이웃 모델

예측하려는 데이터 X가 주어지면 기존 데이터 속성 중 속성이 비슷한 K개의 이웃을 먼저 찾는 알고리즘을 사용한다.
새로운 데이터와 가장 가까운 이웃 데이터를 K개 찾고, 결과를 도출한다.
직선 거리상 가장 가까운 이웃 데이터의 값이 결과로 사용된다.
'''

from sklearn.neighbors import KNeighborsClassifier as kn  

# k 개의 이웃을 설정하는 것, n_neighbors=7 은 가장 가까운 7개의 이웃을 보고 어느 분류에 속하는지 결정하겠다는 뜻
#knn = kn(n_neighbors=7)
knn = kn(n_neighbors=8)

# 훈련
knn.fit(X_train, y_train)

# 테스트
y_knn_pred = knn.predict(X_test)


print("예측값: " , y_knn_pred[:10])
print("실제값: " , y_test[:10].values)
```

```python
from sklearn.metrics import accuracy_score

# accuracy_score() 함수는 정확도를 측정하는 함수이다.
knn_acc = accuracy_score(y_test, y_knn_pred)
print('정확도: ', knn_acc)

# 선생님 정확도 왜 1.0 나오는지..? -> 위에서 random_state를 20으로 해서.. 
```

```python
# 이웃의 숫자를 조절하면서 최적의 정확도를 찾아본다.

k_values = range(1, 21) # 고려할 k값 범위

accuracy_list = [] # 각 k값에 대한 정확도를 저장할 리스트

for k in k_values:
    #print(k)
    knn = kn(n_neighbors=k)
    # 학습을 시켜보자 -> fit()
    knn.fit(X_train, y_train)

    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    accuracy_list.append(accuracy)

accuracy_list

# 결과 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(k_values, accuracy_list, marker='o')
plt.title("Accuracy Score by k Value")
plt.xlabel('k value')
plt.ylabel('Accuracy Score')
plt.grid(True)
plt.show()
```

# SVM 모델

```python
'''
SVN(Support Vector Machine)
데이터 셋의 피쳐로 벡터를 만든다. (벡터를 그래프에 표현하면 하나의 점으로 생각하면 된다.)
각 벡터들이 고유의 축을 갖는 벡터 공간을 이룬하고 가정한다.
모든 데이터를 벡터 공간 내의 좌표에 점으로 표기하고 각 데이터가 속하는 목표 클래스 별로 군집을 이룬다고 생각한다.
각 군집까지의 거리를 최대한 멀리 유지하는 경계면(초평면)을 찾고, 군집을 확연하게 구분한다.
'''

from sklearn.svm import SVC

svc = SVC(kernel='rbf')
svc.fit(X_train, y_train)
```

```python
y_svc_pred = svc.predict(X_test)

print('예측값: ', y_svc_pred[:10])
print('실제값: ', y_test[:10].values)

svc_acc = accuracy_score(y_test, y_svc_pred)
print(svc_acc)
```

# 로지스틱 회귀

```python
# 로지스틱 회귀는 분류문제에 주로 사용되는 모델 (이름에 회귀가 들어가지만 분류 알고리즘을 사용)
# 선형회귀는 선형방정식을 사용하지만 로지스틱 회귀는 시그모이 함수를 사용한다.
# 시그모이 함수는 0~1 사이의 결과만 반환하며, 로지스틱 회귀는 이 결과로 데이터를 0 또는 1로 분류한다. 

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_pred = lr.predict(X_test)

print('예측값: ', y_lr_pred[:10])
print('실제값: ', y_test[:10].values)

lr_acc = accuracy_score(y_test, y_lr_pred)
print("정확도: ", lr_acc) #lr_acc 이거는 함수가 아니다. 변수!!!
```
