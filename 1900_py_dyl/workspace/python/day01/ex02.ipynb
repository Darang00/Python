```python
# requests 모듈
# 파이썬으로 요청을 쉽게 만들어서 보내는 기능을 가지고 있다.

# 설치
!pip install requests 
```

```python
import requests as req

# 요청을 보내고 싶은 주소를 저장해 둔다.
url = 'https://www.naver.com/'

# req.get() : 요청을 보내주는 기능이다.
# 요청을 보내면 HTML코드가 응답으로 반환된다.
resp = req.get(url) 

# resp를 그냥 출력하면 상태코드가 출력된다. 
resp # <Response [200]> :정상 통신 >> 크롤링 가능!

#shift + enter : 본 셀을 실행한 후 새로운 셀을 만들어냄
```

```python
# 상태코드를 확인한다. 200번대는 정상적으로 통신된 것
resp.status_code #200

# 언어설정 (UTF-8: 한글이 잘 나온다.)
resp.encoding #'UTF-8'

# 응답으로 온 데이터 (html.. 사람이 읽기 좋은 형태)
resp.text # <!doctype html> <html lang="ko" class="fzoom"> <head> <meta charset="utf-8"> <meta name="Referrer" content="origin"> <meta http-equiv="X-UA-Compatible" content="IE=edge">

# 응답으로 온 코드 (바이트코드, 컴퓨터가 처리하기 쉬운 형태.. html코드로 응답이 오는데, 태그들 안에 실질적인 값이 들어있는 코드를 반환) 
resp.content #b' <!doctype html> <html lang="ko" class="fzoom"> <head> <meta charset="utf-8"> <meta name="Referrer" content="origin"> <meta http-equiv="X-UA-Compatible
```

```python
# BeautifulSoup4
# HTML 코드를 해석해서 필요한 내용만 추출하는 패키지
# 여러 기능을 제공하며, CSS 선택자를 이용하여 원하는 데이터를 뽑을 수 있다.

# 설치
!pip install BeautifulSoup4
```

```python
from bs4 import BeautifulSoup as bs

# 응답으로 받은 html 코드를 변수에 저장
html = resp.text

# BeautifulSoup 객체로 변환한다.
# BeautifulSoup 객체로 변환을 해야 우리가 파싱하여 원하는 데이터만 뽑을 수 있다.
soup = bs(html, 'html.parser')

# 사람이 보기 쉽게 변환만 했을 뿐 가진 데이터는 변함이 없다.
soup
```

```python
# find('태그명') : 전체 데이터 중 원하는 데이터를 태그 이름으로 가져온다.
# 만약 동일한 태그가 여러개라면 가장 처음 것만 가져온다.
strong = soup.find('strong') #strong 이라는 태그를 찾아라

strong #<strong class="tit">최근 검색어</strong>

# 가져온태그.text : 해당 태그에서 텍스트 데이터만 가져온다.
strong.text #'최근 검색어'
```

```python
# find_all(): 해당 태그를 모두 다 가져온다. 리스트형태로 가져오는것 같다..?
tags = soup.find_all('strong') #[<strong class="tit">최근 검색어</strong>, <strong class="tit">컨텍스트 자동완성</strong>]

tags

for tag in tags: # tags가 list형태니까 for 문으로 꺼내올 수 있음
    print(tag.text)
```

```python
# select('선택자') : css 선택자를 활용하여 원하는 태그를 전부 가져올 수 있다.
# css 선택자
# 특정 id를 가진 태그를 선택: #id명
# 특정 class를 가진 태그를 선택: .class명
titList = soup.select('.tit') #클래스는 '.' 으로 가져올 수 있음 -> .tit : tit라는 클래스를 가진 태그들을 전부 다 가져옴

titList #[<strong class="tit">최근 검색어</strong>, <strong class="tit">컨텍스트 자동완성</strong>]

logo = soup.select('#special-logo')

logo
```

```python
# https://sports.news.naver.com/kbaseball/record/index?category=kbo

url = 'https://sports.news.naver.com/kbaseball/record/index?category=kbo'

resp = req.get(url)

# 상태코드 확인
resp #<Response [200]>

# html 코드 불러오기
html = resp.text

html

# bs(beautifulSoupo) 타입으로 변경
soup = bs(html, 'html.parser')
soup

# 태그의 규칙성을 찾고 해당 규칙으로 태그를 불러온다.
tm_list = soup.select('.tm')

#특정 태그 안에 있는 것을 가져오고 싶으면 > 를 사용한다.
tm_list[0].select('div > span') #div 태그 안의 ('>') span 태그만 가져와라

#span이 여러개일 때 특정 클래스를 제외한 span만 가져오고 싶으면 :not(.class) 를 사용한다.
tm_list[0].select('div > span:not(.emblem)')

result = tm_list[0].select('div > span:not(.emblem)')
result # [<span id="team_LG">LG</span>] : list 형태
result[0].text #'LG'


team_list = []
for tm in tm_list:
    tmp = tm.select('div > span:not(.emblem)')[0]
    team_list.append(tmp.text)

team_list #['LG', 'KT', 'SSG', 'NC', '두산', 'KIA', '롯데', '삼성', '한화', '키움']
```
